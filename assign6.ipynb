```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbs

# Optional style (use one that works in your environment)
plt.style.use('ggplot')

# Load the dataset with semicolon as the delimiter
df = pd.read_csv('student-mat.csv', delimiter=';')

# Print the actual column names to check for correctness


# Display the first few rows to verify the data
print(df.head())
print(df.info())

# Select the correct columns based on the printed column names
# Use 'traveltime' and 'studytime' which are confirmed in the dataset
dfn = df[['traveltime', 'studytime']]  # Make sure these names match exactly

# Display the first few rows of the selected columns
print(dfn.head())

# Assign 'traveltime' and 'studytime' to variables
x = dfn['traveltime']
y = dfn['studytime']

# Plot the lineplot
sbs.lineplot(x=x, y=y, dashes=True)
plt.show()

# Calculate the correlation coefficient and plot heatmap
correlation = dfn.corr()
print("Correlation Matrix:\n", correlation)
sbs.heatmap(correlation, cmap="YlGnBu", annot=True)
plt.show()

# Calculate the covariance and plot heatmap
covar = dfn.cov()
print("Covariance Matrix:\n", covar)
sbs.heatmap(covar, cmap='YlGnBu', annot=True)
plt.show()
```
