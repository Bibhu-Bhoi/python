```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Create a simpple dataset
data={
    'Feature1':[20,30,50,70,90],
    'Feature2':[10,25,35,50,65],
    'Feature3':[100,200,300,400,500]
    }


df=pd.DataFrame(data)

#min-max normalization

def min_max_scaling(df):
    return(df-df.mean())/df.std()
def z_score_scaling(df):
    return(df-df.mean()/df.std())
def decimal_scaling(df):
    max_abs=df.abs().max()
    scaling_factor=10**np.ceil(np.log10(max_abs))
    return df/scaling_factor

min_max_scaled=min_max_scaling(df)
z_score_scaled=z_score_scaling(df)
decimal_scaled=decimal_scaling(df)

print("\nOriginal Dataset:")
print(df)
print("nMin-Max Scaled Data:")
print(min_max_scaled)
print("\nDecimal Scaled Data:")
print(decimal_scaled)

plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
sns.lineplot(data=df)
plt.title('Original Data')
plt.xlabel('Index')
plt.ylabel('Value')

plt.subplot(2,2,2)
sns.lineplot(data=min_max_scaled)
plt.title('Min-Max Scaled Data')
plt.xlabel('Index')
plt.ylabel('Max-Min Value')

plt.subplot(2,2,3)
sns.lineplot(data=z_score_scaled)
plt.title('Z-Score Data')
plt.xlabel('Index')
plt.ylabel('Z-Score Value')

plt.subplot(2,2,4)
sns.lineplot(data=decimal_scaled)
plt.title('Decimal Scaled Data')
plt.xlabel('Index')
plt.ylabel('Scaled Value')

plt.tight_layout()
plt.show()

```
