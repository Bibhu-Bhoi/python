```python
import pandas as pd
import numpy as np
import seaborn as sbs
import matplotlib.pyplot as plt
from scipy.spatial import distance  # Import the distance module

# Load the dataset from the specified path
df = pd.read_csv("student-mat.csv", sep=';')

# Display the column names to identify the correct ones
print(df.columns)

# Select relevant columns
dfs = df[['schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet', 'romantic']]
print(dfs.head())

# Check the structure of the DataFrame
print(df.info())

# Convert 'yes' and 'no' to 1 and 0 respectively
dfs = dfs.replace({'no': 0, 'yes': 1})
print(dfs.head())

# Verify the data type conversion
print(dfs.info())

# Extract arrays for distance calculation
n = np.array(dfs[['schoolsup', 'famsup']])
m = np.array(dfs[['romantic', 'internet']])

# Verify the shape of the arrays
print(n.shape)
print(m.shape)

# Calculate the distance matrix between rows of `n` and `m`
dist_matrix = distance.cdist(n, m, metric='euclidean')

# Check the shape of the distance matrix
print("\nCheck the shape of the distance matrix\n")
print(dist_matrix.shape)
print("\nDistance matrix is:\n")
print(dist_matrix)

# Plot the heatmap of the distance matrix
print("\nHeatmap of the distance matrix\n")
sbs.heatmap(dist_matrix, cmap='viridis')
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sbs
import matplotlib.pyplot as plt
from scipy.spatial import distance

df = pd.read_csv("student-mat.csv", sep=';')

numeric = df[['age', 'Medu', 'Fedu', 'traveltime', 'studytime', 'failures']]
print(numeric.head())  

num1 = np.array(numeric[['age', 'failures']])
print(num1.shape)

num2 = np.array(numeric[['Fedu', 'Medu']])
print(num2.shape)  

dist_matrix = distance.cdist(num1, num2)

print(dist_matrix.shape)

print("\nThe distance matrix is \n")
# print(dist_matrix)

print("\nthe heatmap for the matrix is \n")
sbs.heatmap(dist_matrix, cmap="YlGnBu")
plt.show()
```
